# Wafer example

```python
wdata = pd.read_csv('waferdata.csv')
wdata.head()
```

```python
seaborn.pairplot(wdata)
```

```python
# model = lme(fixed = thickness ~ 1,data=wdata,random= ~ 1 |wafer)
model = smf.mixedlm("thickness ~ 1", wdata, groups=wdata["wafer"]).fit()
print(model.summary())
```
